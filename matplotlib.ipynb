**Format Strings fmt**
You can also use the shortcut string notation parameter to specify the
marker.This parameter is also called fmt, and is written with this syntax:
# marker|line|color

```python
import matplotlib.pyplot as plt
import numpy as np
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, 'o:r')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
ypoints = np.array([3, 8, 1, 10])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.plot(ypoints, marker='o', ms = 20, mec = 'b', mfc = 'r', color='g',linewidth = '20.5',alpha = 0.8)

plt.xlabel("Average Pulse", fontdict = font1)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.grid(color = 'green', linestyle = '--', linewidth = 0.5) # Line Properties for the Grid like color, linestyle, linewidth
plt.show()
```

```python
#multiple line
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])
plt.plot(y1)
plt.plot(y2)
plt.show()
```

```python
import numpy as np

x = np.linspace(0, 4*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.figure(figsize=(10,6))
plt.plot(x, y1, label='Sin(x)', linestyle='-', color='blue', linewidth=2, marker='o', markersize=8)
plt.plot(x, y2, label='Cos(x)', linestyle='--', color='green', linewidth=2, marker='s', markersize=8)

plt.axhline(y=0, color='green', linestyle='--', linewidth=1)
plt.legend(loc='upper right', shadow=True)

plt.grid(True, linestyle='--', alpha=0.2)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Customized Line Plot')

plt.savefig('custom_line_plot.png')

plt.show()
```

**MATPLOTLIB BAR PLOT**

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.bar(x,y, color = 'hotpink', width = 0.5)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bar Plot')

plt.show()
```

```python
categories = ['AI', 'Data Science', 'ML', 'DeepML']
values = [20, 35, 30, 55]

fig, ax = plt.subplots()
bars = ax.bar(categories, values, color='orange')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 1), ha='center', va='bottom')

plt.xlabel('Categories')
plt.ylabel('Quantity')
plt.title('Bar Plot with Quantity and Values')

plt.show()
```

**MATPLOTLIB SCATTER PLOT**

```python
import matplotlib.pyplot as plt

# dataset-1
x1 = [89, 43, 36, 36, 95, 10, 66, 34, 38, 20]
y1 = [21, 46, 3, 35, 67, 95, 53, 72, 58, 10]

# dataset2
x2 = [26, 29, 48, 64, 6, 5, 36, 66, 72, 40]
y2 = [26, 34, 90, 33, 38, 20, 56, 2, 47, 15]

plt.scatter(x1, y1, c ="pink", linewidths = 2, marker ="s", edgecolor ="green", s = 50)

plt.scatter(x2, y2, c ="yellow",linewidths = 2,marker ="^", edgecolor ="red", s = 200)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

**SINE CURVE**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
plt.plot(x, y)

plt.axhline(y=0, color='black', linestyle='--', linewidth=1)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sinusoidal Curve with Horizontal Line at y=0')
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 2 * np.pi, 0.1)
y = np.sin(x)
plt.plot(x, y)

plt.axhline(y=0, color='green', linestyle='--', linewidth=1)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sinusoidal Curve with Horizontal Line at y=0')
```

**MATPLOTLIB PIE CHART**

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
plt.pie(y,labels = mylabels, explode = myexplode, startangle = 90, shadow = True)
plt.legend(title = "Four Fruits:", loc = "lower left")
plt.show()
```

**Seaborn violin plots:** A violin plot is a type of data visualization that combines aspects of both box plots and density plots. It displays a density estimate of the data, usually smoothed by a kernel density estimator, along with the interquartile range (IQR) and median in a box plot-like form.

```python
import seaborn as sns

# load the iris dataset from Seaborn
iris = sns.load_dataset("iris")

# create a violin plot of petal length by species
sns.violinplot(x="species", y="petal_length", data=iris)

# display the plot
plt.show()
```

**SEABORN HEATMAP:** A heatmap is a graphical representation of data that uses colors to depict the value of a variable in a two-dimensional space. Heatmaps are commonly used to visualize the correlation between different variables in a dataset.

```python
import matplotlib as plt
import seaborn as sns
import pandas as pd

path = '/content/drive/MyDrive/datasets/Book5.csv'
df = pd.read_csv(path)

y = df.drop("day",axis=1)
sns.heatmap(y)
```

**Seaborn density plots:**
Density plots, also known as kernel density plots, are a type of data visualization that display the distribution of a continuous variable. They are similar to histograms, but instead of representing the data as bars, density plots use a smooth curve to estimate the density of the data. In Seaborn, density plots can be created using the kdeplot() function.

```python
import seaborn as sns

tips = sns.load_dataset("tips")
tips
```

```python
sns.kdeplot(data=tips, x="total_bill")
```
